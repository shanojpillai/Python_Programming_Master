```python
import matplotlib.pyplot as plt

# Experience years and corresponding roles
experience_years = [1, 3, 5, 8, 12, 15]
roles = ['Technical Architect', 'Solution Architect', 'Domain Architect', 'Business Architect', 'Enterprise Architect', 'Chief Enterprise Architect']

# Convert roles to a numerical axis
role_indices = range(len(roles))

plt.figure(figsize=(10, 6))
plt.plot(experience_years, role_indices, marker='o', linestyle='-', color='b')

# Annotate each point with the role name
for i, role in enumerate(roles):
    plt.text(experience_years[i], i, f" {role}", verticalalignment='bottom')

plt.yticks(role_indices, roles)  # Set text labels on the y-axis
plt.xlabel('Years of Experience')
plt.ylabel('Architectural Roles')
plt.title('Architect Career Progression by Experience')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Experience years and corresponding roles
experience_years = np.array([1, 3, 5, 8, 12, 15])
roles = ['Technical Architect', 'Solution Architect', 'Domain Architect', 'Business Architect', 'Enterprise Architect', 'Chief Enterprise Architect']
role_indices = np.arange(len(roles))

# Set a color map
colors = plt.cm.viridis(np.linspace(0, 1, len(roles)))

plt.figure(figsize=(12, 7))
# Create a scatter plot
scatter = plt.scatter(experience_years, role_indices, s=200, c=colors, alpha=0.6, edgecolors='w', linewidth=2)
# Draw lines
for i in range(len(roles)-1):
    plt.plot(experience_years[i:i+2], role_indices[i:i+2], 'k-', alpha=0.2)

# Annotate each point with the role name
for i, role in enumerate(roles):
    plt.text(experience_years[i] + 0.2, i, role, fontsize=10, verticalalignment='center')

plt.yticks(role_indices, roles, fontsize=12)  # Set text labels on the y-axis
plt.xticks(fontsize=12)
plt.xlabel('Years of Experience', fontsize=14)
plt.ylabel('Architectural Roles', fontsize=14)
plt.title('Architect Career Progression by Experience', fontsize=16, fontweight='bold')
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

# Add a color bar
cbar = plt.colorbar(scatter, shrink=0.5, aspect=5)
cbar.set_label('Progression', rotation=270, labelpad=15)

plt.tight_layout()
plt.show()
```
